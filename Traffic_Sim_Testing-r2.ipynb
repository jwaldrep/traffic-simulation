```python
from car import Car
from road import Road
from simulation import Simulation
import numpy as np
from unittest import mock
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
road = Road(length=1000)
def make_cars(num=10):
    cars = []
    for i in range(num, 0, -1):
        car = Car(road, position=i*10) # reverse order required here...
        cars.append(car)
    return cars
```

```python
cars = make_cars(30)
```

```python
sim = Simulation(cars=cars, steps=100)

sim.start()
positions, speeds = sim.run()
```

```python
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.title("Ten Cars - Dot Size Indicates Speed")
for i in range(0, len(cars), 2):
    plt.scatter(range(len(positions)), positions[:,i])#, speeds[:,i])
    plt.scatter(range(len(positions)), positions[:,i+1], c='g')#, speeds[:,i+1], c='g')
plt.subplots_adjust(left=2, right=4)
```

Took out leapfrogging check in code -- now cars do appear to be able to leapfrog each other (see all the adjacent blues and greens above). This is with 30 cars on a 1000m road. With A 10km road, the cars didn't leapfrog, and all achieved ideal speed (minus occasional random deceleration).
